```python
import numpy as np
import plotly
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

X = np.random.uniform(0, 1, 1000)
#X = np.sort(x)
X = X.reshape(-1, 1)
X
```

```python
w, b = 5, 8

y = w * X + b 
y = y.flatten()
print(y.shape, y)
```

```python
y = y + np.random.normal(loc=0, scale=0.1, size=len(y)) 
y
```

```python
fig = px.scatter(x=X.flatten(), y=y, width=800, height=600)
fig
```

```python

x_flat = X.flatten()
#   w^  b^ 
x_mean, y_mean = np.mean(x_flat), np.mean(y)
w_hat = np.sum((x_flat - x_mean) * (y - y_mean)) / np.sum((x_flat - x_mean) ** 2)
b_hat = y_mean - w_hat * x_mean

print(f": w={w}, b={b}")
print(f":        w^={w_hat:.3f}, b^={b_hat:.3f}")

x_line = np.linspace(np.min(X), np.max(X), 100)
y_theoretical = w * x_line + b
y_hat = w_hat * x_line + b_hat

fig = go.Figure()

#  
fig.add_trace(go.Scatter(x=X.flatten(), y=y, mode='markers', name='',
                         marker=dict(color='blue', size=6)))

#  
fig.add_trace(go.Scatter(x=x_line, y=y_theoretical, mode='lines',
                         name='  (w, b)',
                         line=dict(color='green', dash='dash')))

#  
fig.add_trace(go.Scatter(x=x_line, y=y_hat, mode='lines',
                         name=' (w^, b^)',
                         line=dict(color='red')))

fig.update_layout(
    title="    ",
    xaxis_title="x",
    yaxis_title="y",
    legend=dict(x=0.02, y=0.98, bordercolor="black", borderwidth=1),
    width=800,
    height=600
)

fig

```

```python
from numpy.linalg import LinAlgError
from sklearn.linear_model import LinearRegression

class numpyLinearRegression:
    def __init__(self):
        self.coef_ = None       # ,       X
        self.intercept_ = None  #  

    def fit(self, X, y):
        """
        X: numpy.ndarray (n_samples, n_features)
        y: numpy.ndarray (n_samples,)
        """
        #        (.  np.linalg)
        #  self.coef_  self.intercept_
        X_bias = np.hstack([X, np.ones((X.shape[0], 1))])

        try:
            theta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y
        except LinAlgError:
            print(" X^T X ,   .")
            return

        self.coef_ = theta[:-1]
        self.intercept_ = theta[-1]

    def predict(self, X):
        """
        X: numpy.ndarray (n_samples, n_features)
        """
        return X @ self.coef_ + self.intercept_


#  
my_model = numpyLinearRegression()
my_model.fit(X, y)
y_pred_my = my_model.predict(X)

print(" :")
print("  w^ =", my_model.coef_)
print("  b^ =", my_model.intercept_)

# sklearn
sk_model = LinearRegression()
sk_model.fit(X, y)
y_pred_sk = sk_model.predict(X)

print("\nsklearn:")
print("  w^ =", sk_model.coef_)
print("  b^ =", sk_model.intercept_)

```
