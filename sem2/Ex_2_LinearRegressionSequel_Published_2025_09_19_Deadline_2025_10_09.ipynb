Один из способов борьбы с переобучением (overfitting-ом) — особая модификация функции потерь, так называемая регуляризация. Определим:

$$L_{\lambda}[(w, b)] = \frac{1}{n}\sum\limits_{i=1}^n \left(y_i - \sum_{j=1}^{d}X_{ij}w_j - b\right)^2 + \alpha\sum\limits_{j=1}^d w_j^2.$$

То есть, к привычной нам среднеквадратичной ошибке добавляется сумма квадратов всех весов (кроме сдвига $b$), умноженная на гиперпараметр $\alpha\in\mathbb{R}$.
